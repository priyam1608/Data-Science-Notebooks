# Linear Regression Model

> * **Linear regression is a data analysis technique that predicts the value of unknown data by using another related and known data value.**
> * **It mathematically models the unknown or dependent variable and the known or independent variable as a linear equation.**

> * **Linear regression is a type of supervised machine learning algorithm that computes the linear relationship between the dependent variable and one or more independent features by fitting a linear equation to observed data.**

When there is only one independent feature, it is known as Simple Linear Regression, and when there are more than one feature, it is known as Multiple Linear Regression.

Similarly, when there is only one dependent variable, it is considered Univariate Linear Regression, while when there are more than one dependent variables, it is known as Multivariate Regression.

### Types of Linear Regression Model

1. Simple Linear Regression
2. Multiple Linear Regression
3. Logistic Regression

# Simple Linear Regression

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn import linear_model
```

```python
df = pd.read_csv("ML Practice Files/Linear Regression/homeprices.csv")
df
```

```python
plt.title("Home Prices")
plt.xlabel("Area (in sqft.)")
plt.ylabel("Price (in US Dollars")
plt.scatter(df.area,df.price,color="red",marker='+')
```

```python
reg = linear_model.LinearRegression()
reg.fit(df[['area']],df.price) # fit() method takes first parameter always 2D Dataframe
```

```python
reg.predict( pd.DataFrame([[3300],[4000]], columns=['area']) ) # predict() method takes dataframe here, to predict on it
```

#### Internal Working behind the scenes

```python
reg.coef_   #Slope
```

```python
reg.intercept_    # Y intercept
```

> y = mx+b

> price->(dependent) = m->(coeff.) * area(independent) + b->(intercept)

```python
price = reg.coef_ * 3300 + reg.intercept_
price
```

##### Line (with min errors) Plotting on graph

```python
plt.title("Home Prices")
plt.xlabel("Area (in sqft.)")
plt.ylabel("Price (in US Dollars")
plt.scatter(df.area,df.price,color="red",marker='+')
plt.plot(df.area,reg.predict(df[['area']]),color="blue")
```

# EXERCISE:-
> Based on the per capita incomes of the past years of Canada, Calculate the per capita income in year 2020

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn import linear_model
%matplotlib inline

incomes = pd.read_csv("ML Practice Files/Linear Regression/canada_per_capita_income.csv")
print( incomes.head(5) ,end="\n\n\n")

plt.title("Canada's Past Year Per Capita Incomes")
plt.xlabel("Year")
plt.ylabel("income (US Dollars)")
plt.scatter( incomes.year , incomes['per capita income (US$)'], marker="+", color="red")

model = linear_model.LinearRegression()
model.fit( incomes[['year']], incomes['per capita income (US$)'] )

print("In year 2024, the per capita income is :-")
print(model.predict( pd.DataFrame( [[2024]], columns=['year'] ) ))

plt.plot( incomes.year, model.predict( incomes[['year']] ), color="green")
```
