#### Parsing Date and time from the chronological data

```python
df=pd.read_csv("Practice Files/weather_data2.csv",parse_dates=["day"])
df
```

```python
#Now type of day is converted from string into TimeStamp instance
df.day[0] , type( df.day[0] )
```

#### Use date as an index using "set_index()" method

```python
# Setting day as an index 
df.set_index("day",inplace=True)
df
```

#### Use of "fillna()" method

```python
new_df = df.fillna(0)
new_df
```

```python
#NOTE
new_df = df.fillna(0,limit=1)
new_df

# limit , specifies that the number of null values in a row are the only rows where NaN are filled.
```

> **Here we can observe, event column is illogical for 0 as a value**

```python
#To Fix , we will pass dictionary 

new_df = df.fillna({
    "temperature":0, 
    "windspeed":0,
    "event":"no event"
})
new_df
```

> **Still, Having 0 in temperature and windspeed columns are not seems like accurate estimates**

> **To fix this, filling methods comes in**

```python
# forward fill , means previous value is copied to next null value
new_df = df.ffill()
new_df

# Axis parameter in "ffill()" and "bfill()" method is to set whether you want to copy values horizontally or vertically(default)
```

```python
# backward fill , means next value is copied to previous null value
new_df = df.bfill()
new_df
```

> **Looks like, the data is good, but still it doesn't seems like an accurate estimate**

> **To fix this, we use "interpolate()"**

#### Use of "interpolate()" method

```python
new_df = df.interpolate()
new_df #Avoid the Warning
```

> **default interpolation method is "Linear Interpolation" in which a mid point of two values decides the approximate extimate value for a null field**

> **But Still, as per the missing dates, this can't be an accurate estimate.**

> **Therefore,we will use "Time" method**

```python
new_df = df.interpolate(method='time')
new_df
```

> Now, we can say that, this is an accurate data

#### Use of "dropna()" method

```python
new_df = df.dropna() # removes all the rows with na values
new_df
```

```python
new_df = df.dropna(how="all") 
new_df

#‘any’ : If any NA values are present, drop that row or column.
#‘all’ : If all values are NA, drop that row or column.
```

```python
new_df = df.dropna(thresh=1) # "thresh" parameter says that, if there is a single data in a row , then dont remove it
new_df
```

#### Handling Missing rows in dataframe

> Here we can see, in day column , some dates are missing

```python
new_dates = pd.date_range("01-01-2017","01-11-2017")
new_index = pd.DatetimeIndex(new_dates)
new_df = df.reindex(new_index)
new_df
```

```python
new_df.interpolate(method="time",inplace=True)
new_df
```
