# Decision Tree
**A decision tree is a type of supervised learning algorithm that is commonly used in machine learning to model and predict outcomes based on input data. It is a tree-like structure where each internal node tests on attribute, each branch corresponds to attribute value and each leaf node represents the final decision or prediction. The decision tree algorithm falls under the category of supervised learning. They can be used to solve both regression and classification problems.**

```python
import pandas as pd 

df = pd.read_csv("ML Practice Files/Decision Tree/salaries.csv")
df
```

```python
x = df.drop(['salary_more_then_100k'],axis="columns")        # Independent Variable
y = df.salary_more_then_100k                                 # Dependent Variable
```

```python
from sklearn.preprocessing import LabelEncoder

le = LabelEncoder()
x['company_n'] = le.fit_transform(df.company)
x['job_n'] = le.fit_transform(df.job)
x['degree_n'] = le.fit_transform(df.degree)

x = x.drop(['company','job','degree'],axis="columns")
x
```

```python
from sklearn import tree

model = tree.DecisionTreeClassifier()
model.fit(x,y)
model.predict(pd.DataFrame([[2,0,0]],columns=x.columns))
```

```python
model.score(x,y)
```

In this file using following columns build a model to predict if person would survive or not,

```python
import pandas as pd 

df2 = pd.read_csv("ML Practice Files/Decision tree/titanic.csv")
df2 = df2.drop(['PassengerId','Name','SibSp','Parch','Ticket','Cabin','Embarked'],axis="columns")
df2.head()
```

```python
from sklearn.preprocessing import LabelEncoder

le = LabelEncoder()
df2['Sex_n'] = le.fit_transform(df2.Sex)
final_df = df2.drop(['Sex'],axis="columns")
final_df.head()
```

```python
x = final_df.drop(['Survived'],axis="columns")             # Independent Varibles
y = final_df.Survived                                      # Dependent variable
```

```python
from sklearn.model_selection import train_test_split

x_train, x_test, y_train, y_test = train_test_split(x,y,test_size=0.008,random_state=10)
x_test
```

```python
from sklearn import tree

model = tree.DecisionTreeClassifier()
model.fit(x_train,y_train)
model.predict(x_test)
```

```python
model.score(x_test,y_test)
```
