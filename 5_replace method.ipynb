```python
import pandas as pd
import numpy as np 

df = pd.read_csv("Practice Files/weather_data3.csv")
df
```

```python
# Passing List in replace() method
new_df = df.replace([-99999,'0'],np.nan)
new_df
```

```python
# Passing Dictionary in replace() method
new_df = df.replace({
    "temperature":[-99999],
    "windspeed":[-99999],
    "event":'0'
},np.nan)

new_df
```

```python
# You can also pass dictionary like this
new_df = df.replace({
    -99999:np.nan,
    '0':"Sunny"
})

new_df
```

#### What if data contains these type of values
> **Then we will use Regex**

```python
new_df = df.replace({
    "temperature":'[A-Za-z]',
    "windspeed":'[A-Za-z]'
},'',regex=True)

new_df
```
