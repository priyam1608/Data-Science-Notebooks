# Gradient Descent ALGORITHM
* **Gradient Descent is known as one of the most commonly used optimization algorithms to train machine learning models by means of minimizing errors between actual and expected results. Further, gradient descent is also used to train Neural Networks.**
* **Gradient descent was initially discovered by "Augustin-Louis Cauchy" in mid of 18th century. Gradient Descent is defined as one of the most commonly used iterative optimization algorithms of machine learning to train the machine learning and deep learning models. It helps in finding the local minimum of a function.**

> In this we find the cost function , also known as Mean Square Error (MSE)

#### **here , y_predicted = mx + b**

> * In this algorithm , we have to iteratively calculate slope and b-intercept with respect to some learning rate.
Also, iteratively calculate Cost function based on changing slope and b-intercept

```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline

def gradient_descent(x, y):
    iterations = 10000            #specified by programmer for ease
    learning_rate = 0.08
    m_curr = b_curr = 0           # current slope and b-intercept
    n = len(x)                    # Size of independent variable list
    cost_threshold = 0.0000001  # Stop when cost is below this value

    for i in range(iterations):
        y_predicted = m_curr * x + b_curr
        cost = (1/n) * sum([num**2 for num in (y - y_predicted)])
        md = -(2/n) * sum(x * (y - y_predicted))  # Gradient for m
        bd = -(2/n) * sum(y - y_predicted)        # Gradient for b
        
        # Update parameters
        m_curr = m_curr - learning_rate * md
        b_curr = b_curr - learning_rate * bd
        
        # Stop when cost is below the threshold
        if cost < cost_threshold:
            print(f"Stopped after {i} iterations as cost < {cost_threshold}")
            break

        plt.figure(1)
        plt.plot(x,y_predicted,color='green')
        print(f"Iteration {i+1}: m = {m_curr}, b = {b_curr}, cost = {cost}")
    
    return m_curr, b_curr

# Example data
x = np.array([1, 2, 3, 4, 5])
y = np.array([5, 7, 9, 11, 13])
m, b = gradient_descent(x, y)
print(f"\nOptimal m: {m}, Optimal b: {b}")
```
