# Random Forest

* **Random Forest is a popular machine learning algorithm that belongs to the supervised learning technique.**
* **It can be used for both Classification and Regression problems in ML.**
* **It is based on the concept of ensemble learning, which is a process of combining multiple algorithms to solve a complex problem and to improve the performance of the model.**

> **"Random Forest is a classifier that contains a number of decision trees on various subsets of the given dataset and takes the average to improve the predictive accuracy of that dataset."**

<h2 style="color:green;text-align:center">Applying Random Forest Model to Evaluate Digits Dataset</h2>

```python
import pandas as pd 
from sklearn.datasets import load_digits
from sklearn.model_selection import train_test_split
from sklearn.ensemble import RandomForestClassifier

digit = load_digits()
print(f"directory of Digits dataset - {dir(digit)}")

plt.gray()
for i in range(3):
    plt.matshow(digit.images[i])
plt.show()

df = pd.DataFrame(digit.data)
df['target'] = digit.target

x = df.drop(['target'],axis="columns")
y = df.target

x_train, x_test, y_train, y_test = train_test_split(x,y,test_size=0.2)
print(f"Split Ratio of dataset is {len(x_train) , len(x_test)}")

model = RandomForestClassifier(n_estimators=100)
model.fit(x_train,y_train)
print("\n Syntax of Random forest Model :- \n")
print("RandomForestClassifier(bootstrap=True, class_weight=None, criterion='gini',\nmax_depth=None, max_features='auto', max_leaf_nodes=None,\nmin_impurity_split=1e-07, min_samples_leaf=1,\nmin_samples_split=2, min_weight_fraction_leaf=0.0,\nn_estimators=20, n_jobs=1, oob_score=False, random_state=None,\nverbose=0, warm_start=False)\n")

score = model.score(x_test,y_test)
print(f"Accuracy of the model = {score}")
```

#### Confusion Matrix

```python
from sklearn.metrics import confusion_matrix

y_predicted = model.predict(x_test)
cm = confusion_matrix(y_test,y_predicted)
cm
```

```python
import matplotlib.pyplot as plt 
import seaborn as sm 

plt.figure(1)
sm.heatmap(cm,annot=True)
plt.xlabel('Predicted')
plt.ylabel('Truth')
```

    1) Measure prediction score using default n_estimators (10)

```python
import pandas as pd 
from sklearn.datasets import load_iris
from sklearn.model_selection import train_test_split
from sklearn.ensemble import RandomForestClassifier

iris = load_iris()
print(f"directory of Iris dataset - {dir(iris)}\n")

iris_df = pd.DataFrame(iris.data,columns=iris.feature_names)
iris_df['target'] = iris.target
iris_df['flower_name'] = iris_df.target.apply(lambda x: iris.target_names[x])
print(iris_df.head())

x = iris_df.drop(['target','flower_name'],axis="columns")
y = iris_df.target

x_train1, x_test1, y_train1, y_test1 = train_test_split(x,y,test_size=0.1,random_state=10)
print(f"\nSplit Ratio of dataset is {len(x_train1) , len(x_test1)}")

model2 = RandomForestClassifier(n_estimators=10)
model2.fit(x_train1,y_train1)
y_predicted2 = model2.predict(x_test1)
score2 = model2.score(x_test1,y_test1)
print(f"\nAccuracy of the model = {score2}")
```
