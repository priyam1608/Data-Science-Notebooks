# Matplotlib for Data Science

> **Matplotlib is a powerful plotting library in Python used for creating static, animated, and interactive visualizations. Matplotlibâ€™s primary purpose is to provide users with the tools and functionality to represent data graphically, making it easier to analyze and understand. It was originally developed by John D. Hunter in 2003 and is now maintained by a large community of developers.**

```python
import matplotlib.pyplot as plt
import numpy as np
%matplotlib inline

x=[1,2,3,4,5,6,7]
y=[50,51,52,48,47,49,46]

plt.xlabel("Day")
plt.ylabel("Temperature")
plt.title("Weather")
plt.plot(x,y)
```

#### Parameters in plot() method

```python
plt.plot(x,y,color='green',linewidth=5,linestyle = "dotted",marker='+',markersize=20, alpha = 0.8)
```

### Axes lables,Legend and Grid

```python
days=[1,2,3,4,5,6,7]
max_t=[50,51,52,48,47,49,46]
min_t=[43,42,40,44,33,35,37]
avg_t=[45,48,48,46,40,42,41]
```

```python
#Label
plt.xlabel("Day")
plt.ylabel("Temperature")
plt.title("Weather")

plt.plot(days,max_t ,label="max")
plt.plot(days,min_t ,label="min")
plt.plot(days,avg_t ,label="avg")

plt.legend(loc="lower left") #legend
plt.grid()   #Grid
```

### Bar Chart

```python
company=['GOOGL','AMZN','MSFT','FB']
revenue=[90,136,89,27]

plt.ylabel("(in bln)")
plt.title("US Company Stocks")
plt.bar(company,revenue)
```

> **But, when we have to show two or more parameters in a bar chart, then we have to plot array values first, so that their positions can be changed, then convert them into original String values**

```python
company=['GOOGL','AMZN','MSFT','FB']
revenue=[90,136,89,27]

xpos = np.arange(len(company)) # A list of index of "company" list is generated

plt.ylabel("(in bln)")
plt.title("US Company Stocks")
plt.bar(xpos,revenue)          # Plot that list into bar chart

plt.xticks(xpos,company)       # Replaces the list with original string valued list
plt.show()                     # Removes extra information and show only graph
```

```python
profit=[40,2,34,12]

plt.ylabel("(in bln)")
plt.title("US Company Stocks")
plt.bar(xpos,profit)

plt.xticks(xpos,company)
plt.show()
```

```python
company=['GOOGL','AMZN','MSFT','FB']
revenue=[90,136,89,27]
profit=[40,2,34,12]

plt.ylabel("(in bln)")
plt.title("US Company Stocks")
# By Adding and Subtracting, we are changing the position of bars in bar chart,and width makes them equally wide
plt.bar(xpos-0.2,revenue,width=0.4,label="Revenue")
plt.bar(xpos+0.2,profit,width=0.4,label="Profit")

plt.xticks(xpos,company)
plt.legend(fontsize="medium")
plt.show()
```

```python
company=['GOOGL','AMZN','MSFT','FB']
revenue=[90,136,89,27]
profit=[40,2,34,12]

ypos = np.arange(len(company))

plt.xlabel("(in bln)")
plt.title("US Company Stocks")
# By Adding and Subtracting, we are changing the position of bars in bar chart,and height makes them equally wide
plt.barh(ypos-0.2,revenue,height=0.4,label="Revenue")
plt.barh(ypos+0.2,profit,height=0.4,label="Profit")

plt.yticks(ypos,company)
plt.legend(fontsize="medium")
plt.show()
```

### Histogram

```python
blood_sugar = [113, 85, 90, 150, 149, 88, 93, 115, 135, 80, 77, 82, 129]
plt.hist(blood_sugar) # by default number of bins is set to 10
```

```python
plt.xlabel("Sugar Level")
plt.ylabel("Number Of Patients")
plt.title("Blood Sugar Chart")

blood_sugar_men = [113, 85, 90, 150, 149, 88, 93, 115, 135, 80, 77, 82, 129]
blood_sugar_women = [67, 98, 89, 120, 133, 150, 84, 69, 89, 79, 120, 112, 100]

plt.hist([blood_sugar_men,blood_sugar_women],bins=5,color=['green','orange'],label=['men','women'],orientation="horizontal",rwidth=0.8)
plt.legend()
```

### Pie Chart

```python
exp_vals = [1400,600,300,410,250]
exp_labels = ["Home Rent","Food","Phone/Internet Bill","Car ","Other Utilities"]
```

```python
plt.title("Home Expenses")
plt.pie(exp_vals,labels=exp_labels,shadow=True,autopct="%0.1f%%",radius=1.5 ,explode=[0,0.3,0,0.2,0])
plt.axis("equal")
plt.show()
```
