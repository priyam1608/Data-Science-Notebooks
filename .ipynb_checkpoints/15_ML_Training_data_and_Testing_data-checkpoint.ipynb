# Split Data into 2 Parts

1) **Training Data**
2) **Testing Data**

>It is because ,when we train our model using **Training Dataset** and predict on **Test Dataset**

>Then, we Compare **predicted values** and **actual values** we have

>Also, evaluates a accuracy **score** of our model

```python
import pandas as pd 
import matplotlib.pyplot as plt

df = pd.read_csv("ML Practice Files/Training and Testing data/carprices.csv")

plt.figure("Table 1")
plt.title("Prices of Car as per Mileage")
plt.xlabel("Mileage")
plt.ylabel("Sell Prices($)")
plt.scatter(df['Mileage'],df['Sell Price($)'])
plt.show()

plt.figure("Table 2")
plt.title("Prices of Car as per age")
plt.xlabel("Age (yrs)")
plt.ylabel("Sell Prices($)")
plt.scatter(df['Age(yrs)'],df['Sell Price($)'])
plt.show()

df
```

```python
from sklearn.model_selection import train_test_split

x = df.drop(['Sell Price($)'],axis="columns")
y = df['Sell Price($)']

trainX,testX,trainY,testY = train_test_split(x,y,test_size=0.2,random_state=8)
# here, test size = 0.2 means 20% of original data will be test dataset and "random_state" ensures no random dataset selection

len(trainX) , len(testX)
```

```python
testX
```

```python
testY
```

```python
from sklearn.linear_model import LinearRegression

model = LinearRegression()
model.fit(trainX,trainY)
model.predict(testX)
```

```python
model.score(testX,testY)     #This evaluates the score on the basis of difference in predicted values and actual values
```
